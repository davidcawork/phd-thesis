```python
import pandas as pd
import matplotlib.pyplot as plt
#!pip install openpyxl
```

Conversión de demographics.xlsx a .csv y gráfico de rangos temporales de los despliegues de nodos

```python
df = pd.read_csv('demographics.csv')
```

```python
df['Start'] = pd.to_datetime(df['Start'])
df['End'] = pd.to_datetime(df['End'])
df.head(23)
```

```python
plt.figure(figsize=(16, 10))

for i, row in df.iterrows():
    plt.plot([row['Start'], row['End']], [i, i])
    plt.text(row['Start'], i, f'{i}', ha='right', fontsize=11)

plt.xlabel('Registro de fecha', fontsize=16)
plt.ylabel('ID Residencia', fontsize=16)
plt.title('Series temporales de los despliegues disponibles en SustDataED',fontsize=16)
plt.xlim(pd.Timestamp('2010-06-20'),pd.Timestamp('2014-04-01'))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.savefig('despliegues_nodos.pdf')
plt.show()
```
